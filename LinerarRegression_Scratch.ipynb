```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([2,4,5])
y= np.array([1.2,2.8,5.3])
plt.scatter(x,y)
plt.show()
```

```python
b0 = 0 # intercept
b1 = 3 # Slop
alpha = 0.001 # learning rate
iterations = 1000 # Number of iteration
error = [] #Record error in each iteration
```

```python
for itr in range(iterations):
  error_cost =0
  cost_b0 = 0
  cost_b1 = 0
  for i in range(len(x)):
    y_pred = (b0 + b1 *  x[i])
    error_cost = error_cost + (y[i] - y_pred)**2
    for j in range(len(x)):
      partial_b0 = -1 * (y[j] - (b0 +b1 *x[j]))
      partial_b1 = -1 * x[j] * y[j] - (b0 +b1 *x[j])

      cost_b0 = cost_b0 + partial_b0
      cost_b1 = cost_b1 + partial_b1
    b0 = b0 + alpha * cost_b0
    b1 = b1 + alpha * cost_b1
    error.append(error_cost/(len(x)*2))
```
